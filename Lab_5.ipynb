```python
sns.boxplot(df_csv['MEDV'],color='green')
plt.title('Median value of owner-occupied homes in $1000s')
plt.xlabel('MEDV')
plt.ylabel('Prive in $1000s')
plt.show()
```

```python
sns.scatterplot(x= 'AGE', y= 'MEDV', data= df_csv, color='red')
plt.title('AGE vs MEDV')
plt.xlabel('AGE')
plt.ylabel('MEDV')
plt.show()
```

```python
sns.scatterplot(x= 'RM', y= 'MEDV', data= df_csv, color='blue')
plt.title('RM vs MEDV')
plt.xlabel('RM')
plt.ylabel('MEDV')
plt.show()
```

```python
sns.scatterplot(x= 'CRIM', y= 'MEDV', data= df_csv, color='green')
plt.title('CRIM vs MEDV')
plt.xlabel('CRIM')
plt.ylabel('MEDV')
plt.show()
```

```python
sns.barplot(x='ZN' , y='MEDV', data= df_csv, color='red')
plt.title('ZN vs MEDV')
```

```python
correlation_matrix = df_csv.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Features')
plt.show()
```
