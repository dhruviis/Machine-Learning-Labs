```python
import matplotlib.pyplot as plt
x = np.arange(0,3*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.xlabel('Variable x')
plt.ylabel('trignometric functtions')
plt.title('sine and cosine')
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.xlabel('Variable x')
plt.ylabel('trignometric functtions')
plt.title('Trignometric_sin Function')
plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.xlabel('Variable x')
plt.ylabel('cos function')
plt.title('Trignometric_cos Function')
plt.show()
```

```python
x= np.arrange(0,100,5)
```

```python
x = np.linspace(0, 10, 100)
x_log = np.linspace(0.01, 10, 100)
y_linear = 2*x + 1
y_power = x**2
y_exponential = np.exp(x)
y_logarithmic = np.log(x_log)
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.plot(x, y_linear)
plt.title("Linear Function")
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(x, y_power, color='green')
plt.title("Power Function")
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(x, y_exponential , color='red')
plt.title("Exponential Function")
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(x_log, y_logarithmic , color='purple')
plt.title("Logarithmic Function")
plt.grid(True)

plt.tight_layout()
plt.show()
```
